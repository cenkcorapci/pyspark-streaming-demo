```python
from obspy import read
```

```python
st = read('http://examples.obspy.org/RJOB_061005_072159.ehz.new')
st.plot()
```

```python
st[0].stats
```

```python
for t in st:
    print(len(t.data))
    print(t.data)
    break
```

```python
import numpy as np
import datetime

interval = float(st[0].stats['endtime'] - st[0].stats['starttime']) / 3600.
for t in np.arange(float(st[0].stats['starttime']), float(st[0].stats['endtime']), interval):
    print(datetime.datetime.utcfromtimestamp(t))
    break
```
